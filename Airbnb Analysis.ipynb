#### Task 5- Data Analysis
#### EDA
#### Univariate analysis

```python
e = df.price < 20000
e.value_counts()
```

```python
x = df[df['price'] < 1500]
```

```python
# identifying outliers
sns.boxplot(x.price)
```

```python
plt.figure()
sns.histplot(data = x, x = 'price')
plt.title('price distribution')
plt.ylabel('frequency ')
plt.show()
```

```python
plt.figure()
sns.histplot(data = x, x = 'availability_365')
plt.title('Availability')
plt.ylabel('frequency ')
plt.show()
```

```python
def bar(x,y,df):
  plt.figure()
  sns.barplot( x = x, y = y, data = df)
```

```python
bar('neighbourhood_group', 'price', x)
```

```python
x.info()
```

```python
bar('bedrooms','price', x)
```

```python
bar('room_type','price', x)
```

```python
#feature engineering
x['price_per_bed'] = x['price']/x['beds']
```

```python
x.groupby('neighbourhood_group')['price'].mean()
```

```python
x.groupby('neighbourhood_group')['price_per_bed'].mean()
```
